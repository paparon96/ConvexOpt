# Import packages

```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Settings / parameters

```python
n = 10
```

# Data 

```python
a1 = np.ones(n)
a2 = np.arange((2*n-1)/2, 0, -1)
```

```python
A = np.stack([a1, a2])
A.shape
```

```python
d = np.array([0, 1])
```

# Optimization

```python
# Define and solve the CVXPY problem.
x = cp.Variable(n)
cost = cp.sum_squares(x)
# Create constraint.
constraints = [A*x == d]
prob = cp.Problem(cp.Minimize(cost), constraints)
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("The optimal x is")
print(x.value)
```

```python
plt.plot(x.value, marker="s")
plt.show()
```
