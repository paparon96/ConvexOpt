# Import packages

```python
import numpy as np
```

# Parameters

```python
beta = np.asarray([1, 2, 3])

n = 100
var = 1

mean = [0, 0]
cov = [[1, 0], [0, 1]]
```

# Data generation process

```python
intercept = np.ones(n).reshape(n, 1)
```

```python
explanatory_vars = np.random.multivariate_normal(mean, cov, size = n)
```

```python
X = np.concatenate((intercept, explanatory_vars), axis=1)

print(X.shape)
```

```python
y = X@beta + np.random.normal(0,1,n)
```

# OLS

```python
betahat = np.linalg.inv(X.T@X)@X.T@y
```

```python
betahat
```

```python
# Compute bias of estimates
beta - betahat
```
