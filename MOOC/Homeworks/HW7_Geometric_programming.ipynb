## Convex optimization problem

```
# variables a1(2) a2(2) a3(2) b1 b2 b3
# fill in your solution code here
a1 = cvx.Variable(shape=(2,))
b1 = cvx.Variable(shape=(1,))
a2 = cvx.Variable(shape=(2,))
b2 = cvx.Variable(shape=(1,))
a3 = cvx.Variable(shape=(2,))
b3 = cvx.Variable(shape=(1,))

constraints = [
  a1.T * X - b1 - cvx.maximum(a2.T * X - b2 ,a3.T * X - b3) >= 0
, a2.T * Y - b2 - cvx.maximum(a3.T * Y - b3,a1.T * Y - b1) >= 0
, a3.T * Z - b3 - cvx.maximum(a1.T * Z - b1,a2.T * Z - b2) >= 0
, a1 + a2 + a3 == 0
, b1 + b2 + b3 == 0
]
obj = cvx.Minimize(0)
prob = cvx.Problem(obj, constraints)
prob.solve(cvx.ECOS)
```
